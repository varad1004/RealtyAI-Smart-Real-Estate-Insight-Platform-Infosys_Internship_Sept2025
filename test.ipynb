```python
# Setup: imports and paths
import pandas as pd
from pathlib import Path

root = Path(r"c:\Users\USER\OneDrive\Desktop\realtyAI")
input_path = root / "test.csv"
output_path = root / "housetest.csv"

# Values to consider as missing (NA)
NA_VALS = [
    "", " ", "NA", "N/A", "na", "null", "NULL", "None", "none", "-", "--", "\t", "\r", "\n"
]

print("Input:", input_path)
print("Output:", output_path)
```

```python
# Load CSV and show initial info
print("Reading:", input_path)
df = pd.read_csv(input_path, na_values=NA_VALS, keep_default_na=True, skipinitialspace=True)
print("Initial shape:", df.shape)
print("Columns (first 10):", list(df.columns)[:10])
df.head(3)
```

```python
# Clean: trim strings, normalize empties, drop all-empty rows/cols
# 1) Trim column headers
df.columns = [str(c).strip() for c in df.columns]

# 2) Trim string/object cells
obj_cols = df.select_dtypes(include=["object"]).columns
if len(obj_cols) > 0:
    df[obj_cols] = df[obj_cols].apply(lambda col: col.astype("string").str.strip())

# 3) Replace remaining empty strings with <NA>
df = df.replace({"": pd.NA})

# 4) Drop columns and rows that are entirely NA
cols_before = df.shape[1]
rows_before = df.shape[0]

df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')

cols_after = df.shape[1]
rows_after = df.shape[0]

print({
    "shape_after": df.shape,
    "dropped_cols": cols_before - cols_after,
    "dropped_rows": rows_before - rows_after,
})
```

```python
# Save cleaned data and quick preview
print("Saving to:", output_path)
df.to_csv(output_path, index=False)

# Confirm file saved and show a small preview
from pathlib import Path
p = Path(output_path)
print("Exists:", p.exists(), "Size (bytes):", p.stat().st_size if p.exists() else None)
print("Final shape:", df.shape)
df.head(5)
```
